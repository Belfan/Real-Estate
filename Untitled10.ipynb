## Preparing data :
 
Nous allons tracer les courbes des variations des différentes variables en fonction du prix de vente afin d'avoir un aperçu des relations entre eux.

*N.B:* D'office, Je n'utiliserais pas la latitude et la longitude car selon moi, elles sont en réalité plus complexes à gérer. En effet, on pourrait établir une relation entre les coordonnées géographiques et le prix de vente, mais en réalité la régression linéaire ne prendra pas en compte que telle batiment est par exemple a coté de la maison blanche et qu'il faut donc qu'elle soit plus chère. 

```python
data=data.drop(['No','X5 latitude','X6 longitude'], axis=1)
```

```python
fig, axs = plt.subplots(2,int((data.shape[1]-1)/2))
plot_x=0
plot_y=0
for i in range(data.shape[1]-1):
    if i==int((data.shape[1]-1)/2):
        plot_x=1
        plot_y=0
    if i<(data.shape[1]-1)/2:
        axs[plot_x,plot_y].scatter(data.iloc[:, i], data["Y house price of unit area"],s=10)
    else:
        axs[plot_x,plot_y].scatter(data.iloc[:, i], data["Y house price of unit area"],s=10)
    plot_y=plot_y+1
    
```
